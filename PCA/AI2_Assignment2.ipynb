```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# New Section

```python
faces = np.loadtxt("faces.dat.txt")
faces = faces.reshape(-1, 64, 64)
```

```python
plt.imshow(faces[99].T)
```

```python
for i in range(len(faces)):
  mean = np.mean(faces[i])
  faces[i] = faces[i] - mean

plt.imshow(faces[99].T)
```

```python
PCA_model = PCA(n_components=400)
PCA_model.fit(faces.reshape(-1, 4096))
plt.grid()
plt.plot(PCA_model.explained_variance_)
plt.title('PCA Eigenvalues')
plt.xlabel('Eigenvalue #')
plt.ylabel('Eigenvalue')
plt.show()
```

```python
print(PCA_model.explained_variance_[399])
```

```python
plt.grid()
plt.plot(PCA_model.components_[0])
plt.show()
```

```python
plt.grid()
plt.plot(PCA_model.components_[1])
plt.show()
```

```python
#10 PRINCIPAL COMPONENTS
local_PCA_model = PCA(n_components=10)
local_PCA_model.fit(faces.reshape(-1, 4096))
local_faces = local_PCA_model.inverse_transform(local_PCA_model.fit_transform(faces.reshape(-1, 4096))).reshape(-1, 64, 64)
plt.imshow(local_faces[99].T)
```
